```python
import pandas as pd
train = pd.read_excel('data/mcdonalds.xlsx')
train.head()
```

```python
trainset = train['상호']
```

```python
from collections import Counter
result = Counter(trainset)
print(result)

for key in result:
    print(key, result[key])

#키 없이 카운트 한 값만 알아내고 싶으면
result = Counter(trainset).values()
print(result)
```
